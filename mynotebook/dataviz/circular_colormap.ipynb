# Polar Colormaps

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## Radial Functions

```python
def radial_function(r):
    return r

npts = 1000
theta = [0, 2*np.pi] 
r = np.linspace(0, 1, npts)
z = radial_function(r)
z_g, theta_g = np.meshgrid(z, theta)

fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111, projection='polar')
ax.grid(False)

# Colormap and Colorbar
norm = mpl.colors.Normalize(np.min(z), np.max(z))
pc = ax.pcolormesh(r, theta, z_g, cmap="magma", norm=norm)
cbar = fig.colorbar(pc)
```

## Axial Functions

```python
def axial_function(theta):
    return np.abs( np.pi - theta )

npts = 1000
theta = np.linspace(0, 2*np.pi, npts)
r = [0, 1] 
z = axial_function(theta)
z_g, r_g = np.meshgrid(z, r)

fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111, projection='polar')
ax.grid(False)
ax.set_yticks([]);

# Colormap and Colorbar
norm = mpl.colors.Normalize(np.min(z), np.max(z))
pc = ax.pcolormesh(theta, r, z_g, cmap="magma", norm=norm)
cbar = fig.colorbar(pc)
```
